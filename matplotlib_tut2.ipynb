```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
gas = pd.read_csv("./data/gas_prices.csv")
gas
```

```python
plt.figure(figsize=(8, 5), dpi=200)

plt.title("Gas Price VS Year", fontweight="bold")
plt.xlabel("Year", fontweight="bold", fontsize=12)
plt.ylabel("Gas Price", fontweight="bold", fontsize=12)

for country in gas.columns:
    if country != "Year":
        plt.plot(gas.Year, gas[country], label=country)


plt.xticks(gas.Year[::3].tolist() + [2010])

plt.legend(title="Country", fontsize=5)

plt.grid()
plt.show()
```

```python
fifa = pd.read_csv("./data/fifa_data.csv", index_col=0)
```

```python
fifa.head()
```

```python
fifa.columns
```

### **Histogram**

```python
bin = np.arange(60, 100, 2)

plt.hist(fifa[fifa.Club.isin(["FC Barcelona", "Juventus", "Real Madrid"])].Overall, bins=bin, color="maroon")

plt.xticks(bin)

plt.title("Overall Frequency", fontweight="bold")
plt.xlabel("Overall Rating", fontweight="bold", fontsize=12)
plt.ylabel("Frequency", fontweight="bold", fontsize=12)

plt.grid()
plt.show()
```

### **Pie Chart**

```python
left = fifa[fifa["Preferred Foot"] == "Left"].shape[0]
left
```

```python
right = fifa[fifa["Preferred Foot"] == "Right"].shape[0]
right
```

```python
plt.pie([left, right], labels=["Left", "Right"], colors=["yellow", "indigo"], autopct="%.2f%%")

plt.title("Left Foot vs Right Foot", fontweight="bold")

plt.show()
```

```python
fifa.Weight = [int(x.strip("lbs")) if type(x) == str else x for x in fifa.Weight]
fifa.Weight
```

```python
light = fifa[fifa.Weight < 125].shape[0]
light
```

```python
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].shape[0]
light_medium
```

```python
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].shape[0]
medium
```

```python
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].shape[0]
medium_heavy
```

```python
heavy = fifa[fifa.Weight > 200].shape[0]
heavy
```

```python
plt.pie([light, light_medium, medium, medium_heavy, heavy], autopct="%.2f%%",
        labels=["Light", "Medium Light", "Medium", "Medium Heavy", "Heavy"], explode=[0.4, 0.2, 0, 0, 0.2], pctdistance=0.8)

plt.title("Weight Distribution (in lbs)", fontweight="bold")
plt.show()
```
