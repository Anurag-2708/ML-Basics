### **.loc[#rows, #columns]**

1. If column not mentioned then all columns are shown
2. To get multiple rows or columns use a list
3. __Note-__ While using slice values last index in inclusive
4. .loc[#rows, #columns] -> uses last inclusive slicing and uses labels -> df.loc[2] = label 2
5. .iloc[#rows, #columns] -> uses last exclusive slicing and uses indexing -> df.iloc[2] = 3rd row

```python
df.loc[[0, 1, 2]]
```

```python
df.loc[[0, 1, 2], ['mpg', 'cylinders']]
```

```python
df.loc[0:5, 'mpg':'weight']
```

```python
df.loc[list(range(0,3)) + list(range(10,13)), ['mpg', 'cylinders']] 
```

### **Sort Values**

```python
df.sort_values("mpg", ascending=False) # default is ascending order ->.sort_values = creates a copy
```

```python
df.sort_values(["mpg", "weight"], ascending=[True, False]) # first sort mpg by ascending then weight by descending
```

```python
df.sort_values("horsepower", ascending=False)
```

### **Filtering**

```python
df[(df.horsepower > 100) & ((df.weight < 3000) | (df.origin.str.lower() == "usa"))] # Filtering
```

##### idxmax()

```python
# Car with max horsepower and weight < 3000, number of cylinders = 6

print(df.loc[df[(df.weight < 3000) & (df.cylinders == 6)].horsepower.idxmax(), "name"])
```

##### isin()

```python
df[df.origin.str.lower().isin(["europe", "japan"]) & (df.cylinders > 4)]
```

##### query()

```python
df.query('origin.str.lower().isin(["europe", "japan"]) and cylinders > 4') # no need of writing df everytime
```

### **Grouping**

```python
df.groupby("model_year").mpg.std() # standard deviation of mpg by year
```

##### agg()

```python
df.groupby(["model_year", "cylinders"]).agg(avg_mpg=("mpg", "mean"), std_hp=("horsepower", "std"))
```

```python
df[df.horsepower > 100].groupby(["model_year", "cylinders"]).agg(avg_mpg=("mpg", "mean"), std_hp=("horsepower", "std"))
# First filter by cars with horsepower > 100
# First group by model_year then group by cylinders
# create columns avg_mpg = average mpg of all cars with same cylinders and model_year
# std_hp = standard deviation of horsepower of all cars with same cylinders and model_year
```
