### **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### **Basic Line Graph**

```python
x = [0, 1, 2, 3, 5]
y = [0, 4, 8, 6, 9]

plt.figure(figsize=(5, 3), dpi=200)

plt.plot(x, y, label="Line", color="Green", linewidth=3, 
         marker='.', markersize=15, markerfacecolor="red", 
         markeredgecolor="blue", markeredgewidth=1.5,
         linestyle="-.")

x2 = np.arange(0, 3, 0.001)
plt.plot(x2[:1000], x2[:1000]**2, color="Red", linestyle="-")
plt.plot(x2[999:], x2[999:]**2, color="Red", linestyle="--")

plt.title("First Graph!!", fontname="Comic Sans MS", fontsize= "14", color="maroon")
plt.xlabel("X - Axis", fontsize= 12)
plt.ylabel("Y - Axis", fontsize= 12)

plt.xticks([0, 1, 2, 3, 4, 5, 6])
plt.yticks([0, 2, 4, 6, 8, 10, 12])

plt.legend()

plt.savefig("first_graph.png", dpi=300)

plt.grid()
plt.show()
```

### **Bar Graph**

```python
labels1 = ["first", "second", "third", "fourth"]
values1 = [20, 12, 17, 16]

labels2 = ["one", "two", "three", "four"]
values2 = [13, 11, 9, 14]

plt.figure(figsize=(5, 3), dpi=200)

bars1 = plt.bar(labels1, values1, color="cyan", label="Bar1")
bars2 = plt.bar(labels2, values2, color="maroon", label="Bar2")

bars1[0].set_hatch('/')
bars1[1].set_hatch('O') # use small o for small circle
bars1[2].set_hatch('*')
bars1[3].set_hatch('x')

patterns = ["\\", "-", "+", "."]
for i in range(0, 4):
    bars2[i].set_hatch(patterns[i])

plt.legend()

plt.title("Bar Graph!!", fontname="Comic Sans MS", fontsize= "12", color="maroon")
plt.xlabel("Labels", fontsize= 8)
plt.ylabel("Values", fontsize= 8)

plt.show()
```
